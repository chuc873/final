# Folium and Mapping

Mapping Community Gardens

```python
import folium
from folium import plugins
import pandas as pd
import requests
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from xml.etree import ElementTree

%matplotlib inline
```

```python
gardens=pd.read_csv('NYC_Greenthumb_Community_Gardens.csv')
```

```python
gardens.head()
```

```python
m = folium.Map([40.7341719,-73.992605], zoom_start=11)
```

```python
latlong=gardens[['Latitude','Longitude']]
```

```python
latlong.head()
```

```python
latlong[latlong.isnull()==True]
```

```python
final=latlong.dropna()
```

```python
final.head()
```

```python
m
```

```python
m = folium.Map(location=[40.7341719,-73.992605])
final.apply(lambda row:folium.Marker(location=[row["Latitude"], 
                                                  row["Longitude"]]).add_to(m),
     axis=1)
```

```python
m
```

```python
m.save(outfile= "gardens.html")
```
